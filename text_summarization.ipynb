```python
import os
from dotenv import load_dotenv
from langchain import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_groq import ChatGroq
from langchain.schema import AIMessage, HumanMessage, SystemMessage
from langchain.chains import LLMChain
from langchain.chains.summarize import load_summarize_chain
from langchain_community.document_loaders import PyPDFLoader
```

```python
load_dotenv()
```

```python
groq_api_key = os.getenv("GROQ_API_KEY")
groq_api_key
```

```python
llm = ChatGroq(groq_api_key=groq_api_key, model_name="Llama3-8b-8192")
llm
```

```python
speech = """
People across the country, involved in government, political, and social activities, are dedicating their time to making the ‘Viksit Bharat Sankalp Yatra’ (Developed India Resolution Journey) successful. Therefore, as a Member of Parliament, it is my responsibility to also contribute my time to this program. So today, I have come here not only as a Member of Parliament but also as your ‘sevak’, ready to participate in this program, just like you.  
In our country, governments have come and gone, numerous schemes have been formulated, discussions have taken place, and big promises have been made. However, my experience and observations have led me to believe that the most critical aspect requiring attention is ensuring that the government’s plans reach the intended beneficiaries without any hassles. For example, if there is a ‘Pradhan Mantri Awas Yojana’ (Prime Minister’s Housing Scheme), then those living in jhuggis and slums should get their houses. They should not need to make repeated visits to government offices for this purpose. The government should reach them. Since you have entrusted me with this responsibility, about four crore families have received their ‘pucca’ houses. However, I have also encountered cases where some people were left out of government benefits. Therefore, I have decided to tour the country again to listen to people’s experiences with government schemes, to understand whether they received the intended benefits, and to ensure that programs are reaching everyone as planned—without them having to pay any bribes. We will get the real picture only if we meet the people directly. In this sense, the ‘Viksit Bharat Sankalp Yatra’ is, in a way, my own examination. I want to hear from you and from people across the country whether what I envisioned, and the work I have been doing, aligns with reality and whether it has reached those for whom it was meant.  
It is crucial to check whether the work that was supposed to happen has indeed taken place. I recently met some individuals who used the Ayushman card to get treatment for serious illnesses. One person had a severe accident and, after using the card, could afford the necessary operation. Today, he is recovering well. When I asked him, he said: “How could I have afforded this treatment? But now, with the Ayushman card, I gathered the courage to undergo the operation. Today I am perfectly fine.” Such stories are blessings to me.  
The bureaucrats who prepare good schemes, expedite the paperwork, and allocate funds also feel satisfied that 50 or 100 people who were supposed to get the funds have received them. Funds meant for a thousand villages have been released. But their job satisfaction peaks when they hear that their work has directly impacted someone’s life positively. When they see the tangible results of their efforts, their enthusiasm multiplies, and they feel truly satisfied. Therefore, the ‘Viksit Bharat Sankalp Yatra’ has had a positive impact on government officers. It has made them more enthusiastic about their work, especially when they witness the tangible benefits reaching the people. Officers now feel proud of their work, saying, “I made a good plan, I created a file, and the intended beneficiaries received the benefits.” When they find that the money has reached a poor widow under the Jeevan Jyoti scheme and it has greatly helped her during her time of crisis, they realize they have done a meaningful job. When a government officer listens to such stories, he feels deeply satisfied.  
There are very few who truly understand the power and impact of the ‘Viksit Bharat Sankalp Yatra’. When I hear people in bureaucratic circles talking about it and expressing their satisfaction, it resonates with me. I have heard stories where someone suddenly received 2 lakh rupees after the death of her husband, or a sister shared how the arrival of gas in her home transformed her life. The most significant aspect is when someone says that the line between rich and poor has vanished. While the slogan ‘Garibi Hatao’ (Remove Poverty) is one thing, the real change happens when a person says, “As soon as the gas stove came into my house, the distinction between poverty and affluence disappeared.”
"""
speech
```

```python
chat_message = [
    SystemMessage(content="You are an expert at summarizing speeches."),
    HumanMessage(content=f"Please, Provide a short and concise summary of the following speech:\n Text:{speech}")
]
chat_message
```

```python
llm.get_num_tokens(speech)
```

```python
llm(messages=chat_message)
```

```python
llm(messages=chat_message).content
```

```python
generic_template = """
Write a summary of the following speech:
Speech: {speech}
Translate the precise summary to {language}
"""
generic_template
```

```python
prompt = PromptTemplate(
    input_variables=["speech", "language"],
    template=generic_template
)
prompt
```

```python
complete_prompt = prompt.format(speech=speech, language="Arabic")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
llm_chain
```

```python
summary = llm_chain.run({"speech": speech, "language": "Arabic"})
summary
```

# Stuff Document Chain Text Summarization

```python
pdf_loader = PyPDFLoader(file_path="apjspeech.pdf")
pdf_loader
```

```python
docs = pdf_loader.load_and_split()
docs
```

```python
Stuff_template = """ 
Write a concise and short summary of the following speech,
Speech: {text}
Summary: 
"""
Stuff_template
```

```python
stuff_prompt = PromptTemplate(input_variables=["text"], template=Stuff_template)
stuff_prompt
```

```python
stuff_chain = load_summarize_chain(
    llm=llm,
    chain_type="stuff",
    prompt=stuff_prompt,
    verbose=True
)
stuff_chain
```

```python
stuff_summary = stuff_chain.run(docs)
stuff_summary
```

# MapReduce Document Chain Text Summarization

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=100)
text_splitter
```

```python
final_documents = text_splitter.split_documents(docs)
final_documents
```

```python
len(final_documents)
```

```python
map_template = """ 
Please, Summarize the following speech,
Speech: {text}
Summary: 
"""
map_template
```

```python
map_prompt = PromptTemplate(input_variables=["text"], template=map_template)
map_prompt
```

```python
combine_template = """ 
Provide the final summary of the entire speech, ensuring that it includes these important points.
Add a motivational title, begin the concise summary with an introduction, and then present the summary in numbered points.
Speech: {text}
Summary: 
"""
combine_template
```

```python
combine_prompt = PromptTemplate(input_variables=["text"], template=combine_template)
combine_prompt
```

```python
map_reduce_chain = load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt,
    combine_prompt=combine_prompt,
    verbose=True
)
map_reduce_chain
```

```python
map_reduce_summary = map_reduce_chain.run(final_documents)
map_reduce_summary
```

# Refine Document Chain Summarization

```python
refine_chain = load_summarize_chain(
    llm = llm,
    chain_type="refine",
    verbose=True
)
refine_chain
```

```python
refine_summary = refine_chain.run(final_documents)
refine_summary
```
